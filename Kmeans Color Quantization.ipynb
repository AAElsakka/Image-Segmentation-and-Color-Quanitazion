```python
# importing libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# config
image_path = 'desktop.jpeg'
n_clusters = 5
```

```python
# showing image
image = plt.imread(image_path) /255

image_width, image_height, n_channels = tuple(image.shape)
original_shape = (image_width, image_height, n_channels)

plt.rcParams["figure.figsize"] = (20,6);
plt.imshow(image)
print(image.shape)
```

```python
from sklearn.cluster import KMeans

image_array = np.reshape(image, (image_width * image_height, n_channels))

kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(image_array)

labels = kmeans.predict(image_array)
```

```python
plt.imshow(labels.reshape(image_width, image_height));
```

```python
plt.rcParams["figure.figsize"] = (20,6);
plt.imshow(kmeans.cluster_centers_[labels].reshape(image_width, image_height, -1));
```
